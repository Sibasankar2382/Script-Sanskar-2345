# Data Collection :

To get the data for Nifty fifty first we get the list of all fifty stock from NSE
and use the symbols of the stocks

```python
# here we are using library yahoo fainance to get the data for each stock
import yfinance as yf
import pandas as pd
```

```python
# NSE csv for stocks list
stocks = pd.read_csv('Nifty_stocks.csv')
stocks.head()
```

```python
stocks.columns = stocks.columns.str.strip()
stocks.head()
```

```python
# by using the pnadas series we are getting the list of the stock from the NSE csv
nifty_stocks = pd.Series(stocks['SYMBOL'][1:])
nifty_stocks
```

```python
# Here have to get the data for last 4 year so we are collecting data from '2020-01-01' to ''2024-01-01''
start_date = '2020-01-01'
end_date = '2024-01-01'

# Storing all stocks data in a dictionary
nifty_data = {}
# by using loop we are iteration trough all stocks data
for stock in nifty_stocks:
    data = yf.download(f'{stock}.NS', start=start_date, end=end_date)
    nifty_data[stock] = data
```

# Exploratory Data Analysis (EDA) :

```python
# we need to import some libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#import chart_studio.plotly as py 
import plotly.graph_objs as go
from plotly.offline import plot

# for offline ploting
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
# Here we are creating the fuction which will take a input of stock name and will do the EDA 
def stock_eda(stock_name):
    print(pd.DataFrame(nifty_data[stock_name].head()))
    print('Dataset Information :')
    print(nifty_data[stock_name].info())
    print()
    print('Checking Null/NaN :')
    print(nifty_data[stock_name].isna().sum())
    print()
    print('Check for any duplicates :')
    print('dulpicates: ', nifty_data[stock_name].duplicated().sum())
    print()
    print('Statistical Analysis :')
    print(pd.DataFrame(nifty_data[stock_name].describe()))
    print()
    print('Visulization graphs :')
    print('Boxplot of', stock_name)
    nifty_data[stock_name][['Open','High','Low','Close','Adj Close']].plot(kind='box')
    layout = go.Layout(title=f'Stock price of {stock_name}',
                      xaxis = dict(title='Date',
                                  titlefont=dict(
                                  family='Courier New, monospace',
                                  size=18, color='#7f7f7f')),
                     yaxis = dict(title='Price',
                                  titlefont=dict(
                                  family='Courier New, monospace',
                                  size=18, color='#7f7f7f')))
    stoc_data = [{'x':nifty_data[stock_name].index, 'y':nifty_data[stock_name]['Close']}]
    plot = go.Figure(data=stoc_data, layout=layout)
    iplot(plot)
```

# Model Creation / Portfolio Analysis :

```python
# to train the model we need to import the following libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import time
import datetime
from sklearn.preprocessing import PolynomialFeatures
import joblib
```

```python
def calculate_returns(stock_name, quantity, buying_date, selling_date,):
    if selling_date <= '2024-01-01':
        # Use historical data
        buying_price = nifty_data[stock_name].loc[buying_date]['Close']
        selling_price = nifty_data[stock_name].loc[selling_date]['Close']
        pnl = (selling_price - buying_price) * quantity
        print('Buying price:', buying_price)
        print('Selling price:', selling_price)
        print('Profit/Loss:', pnl)
    else:
    
    # Predict prices for the buying and selling dates
        y = nifty_data[stock_name]['Close']
        x = nifty_data[stock_name][['Year', 'Month', 'Day','Volume']]
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
        poly = PolynomialFeatures(degree=2) # Polynomial regression can capture more complex relationships between the independent and dependent variables
        x_train_poly = poly.fit_transform(x_train)
        x_test_poly = poly.transform(x_test)
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
        poly = PolynomialFeatures(degree=2)
        x_train_poly = poly.fit_transform(x_train)
        x_test_poly = poly.transform(x_test)
        m = joblib.load('stock_model.joblib')
        m.fit(x_train_poly, y_train)
        y_pred=m.predict(x_test_poly)
        y_predt=m.predict(x_train_poly)
        
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test, y_pred)
        n = len(y_test)
        p = x_test.shape[1]
        adj_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
        
        print("R-squared score:", r2)
        print("Adjusted R-squared score:", adj_r2)
        
        buying_date = datetime.datetime.strptime(buying_date, '%Y-%m-%d').date()
        selling_date = datetime.datetime.strptime(selling_date, '%Y-%m-%d').date()
        
        last_100_data = data.iloc[-100:]
        volume_mean = np.mean(last_100_data['Close'])
        volume_std = np.std(last_100_data['Close'])

        buying_year = buying_date.year
        buying_month = buying_date.month
        buying_day = buying_date.day
        selling_year = selling_date.year
        selling_month = selling_date.month
        selling_day = selling_date.day

        # Create feature vectors for buying and selling dates
        X_buying = np.array([[volume_mean,buying_year,buying_month,buying_day ]])
        X_buying_scaled = (X_buying - [volume_mean, 0, 0, 0]) / [volume_std, 1, 1, 1]
        predicted_close_buying = m.predict(poly.transform(X_buying_scaled))[-1]

        X_selling = np.array([[volume_mean,selling_year,selling_month,selling_day ]])
        X_selling_scaled = (X_selling - [volume_mean, 0, 0, 0]) / [volume_std, 1, 1, 1]
        predicted_close_selling = m.predict(poly.transform(X_selling_scaled))[-1]

        # Calculate profit or loss
        buying_price = predicted_close_buying
        selling_price = predicted_close_selling
        pnl = (selling_price - buying_price) * quantity

        # Display the results
        print('Predicted Buying price:', predicted_close_buying)
        print('Predicted Selling price:', predicted_close_selling)
        print('Profit/Loss:', pnl)
        
```
